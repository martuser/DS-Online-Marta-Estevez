#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_California_cities.head()
```

```python
# 1.2.1: 1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  
top5 = df_California_cities.nlargest(5, "population_total").copy()

# En la figura la poblacin est en miles
top5["Pop_miles"] = top5["population_total"] / 1000

colores = ["mediumslateblue", "gray", "salmon", "darkkhaki", "rosybrown"]  # Colres buscados aqui: https://matplotlib.org/stable/gallery/color/named_colors.html, muy guay

plt.figure(figsize=(10, 8))
plt.bar(top5["city"], top5["Pop_miles"], color=colores) 

plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.ylim(0, 5000)

# nmeros encima de cada barra (centrados con ha="center")
for i, v in enumerate(top5["Pop_miles"].values):
    etiqueta = f"{v:.1f}" if float(v).is_integer() else f"{v:.3f}"
    plt.text(i, v + 80, etiqueta, ha="center")

plt.show()
```

```python
#1.2.2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
plt.figure(figsize=(10, 8))
plt.hist(df_California_cities["area_total_km2"], bins=50, color="steelblue", label="Hist Area")

font_dict = {
    "size" : 18,
    "family": "Bradley Hand ITC",
}
plt.xlabel("Superficie ($km^2$)", fontdict = font_dict)
plt.ylabel("Nmero de ciudades", fontdict = font_dict)
plt.legend()

plt.show()
```

```python
#1.2.3. El scatter de rea de agua de un ciudad con su poblacion.
plt.figure(figsize=(10, 8))

x = df_California_cities["area_water_km2"]
y = df_California_cities["population_total"] / 1000  # miles de habitantes

plt.scatter(x, y, color="green", alpha=0.6)
plt.title("Poblacin por kilometro cuadrado de agua")
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles habitantes)")

# Dos puntos outliers
out = df_California_cities.nlargest(2, "area_water_km2").copy()
out_x = out["area_water_km2"].values
out_y = (out["population_total"].values / 1000)

arrows = {
    "arrowstyle": "->"
}
# Texto + 2 flechas
plt.annotate("Outliers",
             xy=(out_x[0], out_y[0]),
             xytext=(60, 3500),
             arrowprops=arrows)
plt.annotate("Outliers",
             xy=(out_x[1], out_y[1]),
             xytext=(60, 3500),
             arrowprops=arrows)

plt.show()
```

```python
#1.2.4. El boxplot de la altura de las ciudades.
plt.figure(figsize=(10, 8))
plt.boxplot(
    df_California_cities["elevation_m"],
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor="red", color="black"),
    medianprops=dict(color="black"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
)

plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)")

plt.show()
```

```python
# 1.2.5 Grfico de todas las figuras en uno
fig, axs = plt.subplots(2, 2, figsize=(18, 18))
# --- (1) ---
top5 = df_California_cities.nlargest(5, "population_total").copy()
top5["Pop_miles"] = top5["population_total"] / 1000
colores = ["mediumslateblue", "gray", "salmon", "darkkhaki", "rosybrown"]  
axs[0, 0].bar(top5["city"], top5["Pop_miles"], color=colores)
axs[0, 0].set_title("Ciudades ms pobladas de California")
axs[0, 0].set_xlabel("Ciudades")
axs[0, 0].set_ylabel("Habitantes (miles)")
axs[0, 0].set_ylim(0, 5000)

for i, v in enumerate(top5["Pop_miles"].values):
    etiqueta = f"{v:.1f}" if float(v).is_integer() else f"{v:.3f}"
    axs[0, 0].text(i, v + 80, etiqueta, ha="center")

# --- (2) ---
font_dict = {
    "size" : 18,
    "family": "Bradley Hand ITC",
}

axs[0, 1].hist(df_California_cities["area_total_km2"], bins=50, color="steelblue", label="Hist Area")
axs[0, 1].set_xlabel("Superficie ($km^2$)", fontdict = font_dict)
axs[0, 1].set_ylabel("Nmero de ciudades", fontdict = font_dict)
axs[0, 1].legend()

# --- (3) ---
x = df_California_cities["area_water_km2"]
y = df_California_cities["population_total"] / 1000 

axs[1, 0].scatter(x, y, color="green", alpha=0.6)
axs[1, 0].set_title("Poblacin por kilometro cuadrado de agua")
axs[1, 0].set_xlabel("rea de agua en la ciudad ($km^2$)")
axs[1, 0].set_ylabel("Poblacin (miles habitantes)")

out = df_California_cities.nlargest(2, "area_water_km2").copy()
out_x = out["area_water_km2"].values
out_y = (out["population_total"].values / 1000)
arrows = {
    "arrowstyle": "->"
}
axs[1, 0].annotate("Outliers",
             xy=(out_x[0], out_y[0]),
             xytext=(60, 3500),
             arrowprops=arrows)
axs[1, 0].annotate("Outliers",
             xy=(out_x[1], out_y[1]),
             xytext=(60, 3500),
             arrowprops=arrows)

# --- (4) ---
axs[1, 1].boxplot(
    df_California_cities["elevation_m"],
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor="red", color="black"),
    medianprops=dict(color="black"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
)

axs[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axs[1, 1].set_xlabel("Altitud (m)")


plt.tight_layout()
plt.show()
```
